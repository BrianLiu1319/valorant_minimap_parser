```python
def load_minimaps():
    directory = "./imgs/"

    minimaps = []
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
        # checking if it is a file
        if os.path.isfile(f):
            minimaps.append(f)
    return minimaps
```

# Minimap Classifier

```python
# Load the images
for i in minimaps:
    img1 = cv.imread('./frames/0.jpg')  # Image 1
    img2 = cv.imread(i)  # Image 2
    # Check if the images are loaded properly
    if img1 is None or img2 is None:
        print("Could not open or find the images.")
        continue
        
    # Initiate SIFT detector
    sift = cv.SIFT_create()
     
    # find the keypoints and descriptors with SIFT
    kp1, des1 = sift.detectAndCompute(img1,None)
    kp2, des2 = sift.detectAndCompute(img2,None)
     
    # BFMatcher with default params
    bf = cv.BFMatcher()
    matches = bf.knnMatch(des1,des2,k=2)
     
    # Apply ratio test
    good = []
    for m,n in matches:
        if m.distance < 0.75*n.distance:
            good.append([m])

    # cv.drawMatchesKnn expects list of lists as matches.
    img3 = cv.drawMatchesKnn(img1,kp1,img2,kp2,good,img1,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    plt.imshow(img3),plt.show()
    print(i + " " + str(len(good)))
```

```python
def minimap_classifier(frame):
    # get frame, return str of likely minimap else return NONE
    minimaps = load_minimaps()

    likely = None
    likely_count = 0
    total = []
    count = 0
    
    for i in minimaps:
        img1 = cv.imread(frame)  # Image 1
        img2 = cv.imread(i)  # Image 2
        # Check if the images are loaded properly
        if img1 is None or img2 is None:
            print("Could not open or find the images.")
            continue

        count += 1
        # Initiate SIFT detector
        sift = cv.SIFT_create()
         
        # find the keypoints and descriptors with SIFT
        kp1, des1 = sift.detectAndCompute(img1,None)
        kp2, des2 = sift.detectAndCompute(img2,None)
         
        # BFMatcher with default params
        bf = cv.BFMatcher()
        matches = bf.knnMatch(des1,des2,k=2)
         
        # Apply ratio test
        good = []
        for m,n in matches:
            if m.distance < 0.75*n.distance:
                good.append([m])

        total.append(len(good))
        if len(good) > 40 and len(good) > likely_count:
            likely = i
            likely_count = len(good)

    total.sort(reverse=True)

    # an image that has a high average is likely not a map
    if (total[0] + total[1] + total[2]) / 3 <= 40: 
        return likely
    else:
        return None
```

# MINIMAP FINDER

```python
def good_matches(frame, template):
    img1 = cv.imread(frame)  # Image 1
    img2 = cv.imread(template)  # Image 2
    # Check if the images are loaded properly
    if img1 is None or img2 is None:
        print("Could not open or find the images.")
        
    # Initiate SIFT detector
    sift = cv.SIFT_create()
     
    # find the keypoints and descriptors with SIFT
    kp1, des1 = sift.detectAndCompute(img1,None)
    kp2, des2 = sift.detectAndCompute(img2,None)
     
    # BFMatcher with default params
    bf = cv.BFMatcher()
    matches = bf.knnMatch(des1,des2,k=2)
     
    # Apply ratio test
    good = []
    for m,n in matches:
        if m.distance < 0.75*n.distance:
            good.append([m])
    
    return good
```

```python
img1 = cv.imread('./frames/0.jpg')  # Image 1
img2 = cv.imread('./imgs/Lotus_minimap.png')  # Image 2
# Check if the images are loaded properly
if img1 is None or img2 is None:
    print("Could not open or find the images.")
    
# Initiate SIFT detector
sift = cv.SIFT_create()
 
# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
 
# BFMatcher with default params
bf = cv.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)
 
# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])

# cv.drawMatchesKnn expects list of lists as matches.
img3 = cv.drawMatchesKnn(img1,kp1,img2,kp2,good,img1,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
```

```python
pts = []
for match in good:
    p1 = kp1[match[0].queryIdx].pt
    pts.append(p1)

dots = img1
```

```python
# get all dots 
for i in pts:
    x = int(i[0])
    y = int(i[1])
    dots = cv.circle(dots, (x,y), radius=2, color=(0, 0, 255), thickness=-1)

plt.imshow(dots),plt.show()
```

```python
# find all points that are matched according to BFmatcher algorithm
# plt.scatter(*zip(*pts))

for i in pts:
    x, y = i
    plt.scatter(x,-y, color="cornflowerblue")

plt.show()
```

## we want to cluster dots, classify them then get the coordinates of rectangle to determine where minimap is


```python
from sklearn.cluster import KMeans

X = np.array(pts)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
# Plot the scatterplot with clusters
plt.scatter(X[:, 0], -X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# # Plot the centroids
# centroids = kmeans.cluster_centers_
# plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75)

plt.title('K-Means Clustering on a Scatter Plot')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.show()
```

```python
# get cluter with the greatest counts
cluster_counts = Counter(y_kmeans)
cluster_label, count = cluster_counts.most_common(1)[0]

# Get the points in the largest cluster
largest_cluster_points = X[y_kmeans == cluster_label]

# Plot all data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], -X[:, 1], c='lightgray', label='Other Clusters')

# Plot only the largest cluster
plt.scatter(largest_cluster_points[:, 0], -largest_cluster_points[:, 1], c='blue', label='Largest Cluster')


plt.title('Largest Cluster with Consistent Axes')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.legend()
plt.show()
```

```python
def load_frames():
    directory = "./frames/"

    frames = []
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
        # checking if it is a file
        if os.path.isfile(f):
            frames.append(f)
    return frames 
```

```python
frame_list = load_frames()

template = "./imgs/Lotus_minimap.png"
good_lst = []

for i in frame_list:
    good_lst += good_matches(i, template)

pts = []
for match in good_lst:
    p1 = kp1[match[0].queryIdx].pt
    pts.append(p1)
```
